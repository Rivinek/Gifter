```python
import random
import string

import pandas as pd
import numpy as np
from nltk.corpus import reuters

from gifter.modeling.llda.llda import LLDA
```

```python
"""Random sample to test"""
random_sample = 100
id_list = random.sample(reuters.fileids(), random_sample)
```

```python
labels = []
corpus = []
for _id in id_list:
    labels.append(reuters.categories(_id))
    corpus.append(map(str.lower, map(str, reuters.words(_id))))
labelset = list(set(reduce(list.__add__, labels)))
```

```python
labelset
```

```python
topics_number = 50
alpha = 0.001
beta = 0.001
llda = LLDA(topics_number, alpha, beta)
llda.set_corpus(labelset, corpus, labels)
```

```python
for i in range(random_sample):
    llda.inference()
print "perplexity : %.4f" % llda.perplexity()
```

```python
phi = llda.phi()
for k, label in enumerate(labelset):
    print "\n-- label %d : %s" % (k, label)
    for w in np.argsort(-phi[k])[:20]:
        print "%s: %.4f" % (llda.vocas[w], phi[k, w])
```

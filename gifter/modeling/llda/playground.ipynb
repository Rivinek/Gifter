```python
import pandas as pd
from gifter.modeling.tokenizer import preprocess, lemmatize

def parse_doc(category, screen_name):
    path = '/home/rivinek/Documents/labeled_twitter/{}/{}'.format(category,
                                                                  screen_name)
    return pd.io.json.read_json(path)
```

```python
tweets = parse_doc('entertainment', 'pre_bobsaget.json')
```

```python
tweets
```

```python
for i in range(100):
        row = tweets.ix[i]
        print "-"*40
        print "Original: ", row['text']
        print "Preprocessed: ", preprocess(row['text'], row['entities'])
        print "Lemmatized: ", lemmatize(
            preprocess(row['text'], row['entities'])
            )
```

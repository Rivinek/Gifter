```python
import pandas as pd
from tokenizer import (
    preprocess,
    lemmatize,
)
%matplotlib inline
```

```python
df = pd.io.json.read_json("./data.json")
%time lemmas = pd.Series([lemmatize(preprocess(*row)) for idx, row in df[['text', 'entities']].iterrows()])
```

```python
%time counts = pd.Series(lemmas.sum()).value_counts()
```

```python
counts.head(30).plot(kind='bar')
```
